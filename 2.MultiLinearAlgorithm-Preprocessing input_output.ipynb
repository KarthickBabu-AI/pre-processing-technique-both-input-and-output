```python
import pandas as pd #Importing Pandas Library to fetch data from the database
import pickle
```

```python
data=pd.read_csv("Startups.csv")#importing the base file to understand the data structure
data
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)#This step to convert categorical data into binary using get.dummies using pandas library
data
```

```python
data.columns #Identify no.of columns to use to assign variable to input and output
```

```python
independent=data[['R&D', 'Administration', 'Marketing', 'State_Karnataka','State_Madurai']] #Assigning varibale to input.
dependent=data[['Profit']] #Assigning varibale to output.
```

```python
from sklearn.model_selection import train_test_split #Splitation of traing and test set
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```
